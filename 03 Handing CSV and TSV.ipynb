## read_csv Arguments 

### Age Group Data Set

```python
# Load the age group csv file 
df = pd.read_csv('AgeGroupDetails.csv')    
df
```

```python
# Load the age group without header 
df = pd.read_csv('AgeGroupDetails_NoHeader.csv')    
df
```

```python
# Load the age group without header , Removing the header
df = pd.read_csv('AgeGroupDetails_NoHeader.csv',header=None)    
df
```

```python
# Renaming or Providing the User defined header names
df =pd.read_csv('AgeGroupDetails_NoHeader.csv', names=['S.No','Age Group','Total Cases','Percentage'])
df
```

```python
# Renaming or Providing the User defined header names and also adding new columns
df =pd.read_csv('AgeGroupDetails_NoHeader.csv', names=['S.No','Age Group','Total Cases','Percentage','State'])
df
```

```python
# Removing the comments or unused rows
df =pd.read_csv('AgeGroupDetails_NoHeader.csv',names=['S.No','Age Group','Total Cases','Percentage'],skiprows = [10,11])
df
```

```python
# Filling particular values by any user defined values 
df =pd.read_csv('AgeGroupDetails_NoHeader.csv',names=['S.No','Age Group','Total Cases','Percentage']
                ,skiprows = [10,11]
                ,na_values=['Missing'])
df
```

### Save the CSV File 

```python
df.to_csv('AgeGroupDetails_Final.csv')
```

```python
# Export the CSV File without Index and Column 
df.to_csv('AgeGroupDetails_Final_NOIndex.csv',index=False,header=False)
```

### 2. Text Delimited files

```python
# Loading the text file (Tab Seperated Value)
df = pd.read_table('AgeGroupDetails.txt')    
df
```

```python
# Loading the text file (Tab Seperated Value)
df = pd.read_table('AgeGroupDetails.txt',sep='\t')    
df
```
